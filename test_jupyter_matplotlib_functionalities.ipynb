```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib
import math
from matplotlib import pyplot as plt
from textwrap import wrap

# draw time series given a univariate normal distribution, a discrete uniform distribution
ts = pd.Series(data=np.random.randn(100), dtype='double')
ts2 = pd.Series(data=np.random.randint(-2,3,100))
# also prepare x for a probability density function for a normal distribution
mu=0
sigma=1
x_normal=np.linspace(mu-2*sigma,mu+2*sigma,100)

fig = plt.figure(figsize=(8,4), dpi=90, facecolor='w', edgecolor='w')
fig.suptitle("Various Plots", fontsize=16,y=2.0)
fig.subplots_adjust(hspace=1, top=1.7)
ax, ax2, ax3 = fig.subplots(3, 1, sharey=False)
ax.set_title("\n".join(wrap("Times Series: univariate normal distribution", 20)))
ax2.set_title("\n".join(wrap("Time Series: discrete uniform distribution",20)))
ax3.set_title("normal distribution")
# in this case, we are plotting the 2 time series in a different subplot
ax.plot(ts, color=[0.0, 1.0, 0.0])
ax2.plot(ts2, color=[0.0, 0.0, 0.1])
ax3.plot(x_normal, stats.norm.pdf(x_normal, mu, sigma))

ax.set_xlabel('time (s)',fontsize='15')
ax.set_ylabel('value', fontsize='20')
ax.xaxis.label.set_color('red')
ax.yaxis.label.set_color('red')
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='blue')
ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')
ax.spines['right'].set_color('blue')
```

* Above draws on the same plot as subplots, 2 time series given a univariate normal distribution and a discrete uniform distribution, and 1 probability density function for a gaussian distribution based on mean=0, and sigma=1
